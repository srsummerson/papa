#### Free-choice probabilistic reward task with stimulation. High value target rewarded with 80% probabilitiy and low value target rewarded with 40% probability. During Block A (100 trials), there are randomly interleaved instructed trials that make up 30% of the block. In today's task, the subject did not learn the probabilities of reward well, as shown by his inflated probability of selecting the low value target in Block A.  In Block B (100 trials), all trials are instructed and stimulation (200 Hz,200 us, 1V) is delivered during the hold period (1000 ms) of instructed trials to the low value target.  In Block A', there are randomly interleaved instructed trials to the low value target which make up 30% of all trials and during the hold period stimulation is delivered.
