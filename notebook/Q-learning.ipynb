### Function to compute the log-likelihood of the probability of all choices to low for a particular alpha and beta paramenter in the standard Temporal Difference reinforcement learning model.

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
def LogLikelihoodRLPerformance(Q_initial,alpha, beta, reward_schedule,choice):
    Q_initial[1] = Q_low
    Q_initial[2] = Q_high
    num_trials = choice.length

    for i in range(0,num_trials):
        Q_low += alpha*(choice[i]==1)*(reward_schedule[i] - Q_low)
        Q_high += alpha*(choice[i]==2)*(reward_schedule[i] - Q_high)
        prob_choice_low[i] = np.exp(beta*Q_low)/(np.exp(beta*Q_low) + np.exp(beta*Q_high))
        prob_total_low = prob_total_low*prob_choice_low[i]
    
    return np.log(prob_total_low)
    
```
